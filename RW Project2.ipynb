# DataSet 1 Rellgion and income

Importing cvs untidy dataset

```python
import pandas as pd
df_relinc=pd.read_csv("relinc.csv")
df_relinc.head()
```

Making CVS tablet into an analyze the data tidy transform. There are three variables that would be used in this dataset is religion, income, and frequency. (Using melt for the long format)

```python

df_relinc=df_relinc.melt(id_vars=["religion"],var_name=["income"],value_name="frequency")
```

```python
df_relinc.head()
```

Adding the import original cvs untidy file with the newly tidy file. Using pivot table to a wide format

```python


df_relinc=(df_relinc.pivot_table(index = "religion", columns = "income", values = "frequency")
   .reset_index() 
   .rename_axis(None, axis = 1))
```

```python
df_relinc.head()
```

Importing the CVS (untidy dataset) Then making it Tiny Dataset then turning it back to (untidy) helps manipulating the data. 

# DataSet weather speeds 2

Importing CVS Weather File Using 5 samples. (this is a very quick way to import the file. and just code later)

```python


import datetime

df_weather = pd.read_csv('weather-raw.csv')

df_weather.sample(5)
```

Since The CVS is now imported. Its time to add the data into Rows ( ID,Year,Month,Element and Temp)

```python

df_weather=df_weather.melt(id_vars=["id","year","month","element"],var_name=["day"],value_name="temp")
df_weather.update(pd.DataFrame({"day":[day[1:] for day in df_weather["day"]]}))

df_weather.sample(5)
```
