# Q1. Load the ratings by user information that you collected into a pandas dataframe. Using custom CSV file

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing

movie_data = pd.read_csv("/Users/bobby/Desktop/movies/movies.csv", index_col=0)
movie_data
```

# Q2a. Show the average ratings for each user and each movie.
Finding the User Average column by performing the mean function

```python
movie_data['User Average'] = movie_data.mean(axis = 1, skipna = True)
movie_data
```

# Q2b. Show the average ratings for each movie.

```python
movie_data.loc['Average Movie Rating'] = movie_data.mean(axis = 0, skipna = True)
movie_data
```

# Q3. Create a new pandas dataframe, with normalized ratings for each user. Again, show the average ratings for each user and each movie.



```python
normalized_data = movie_data.copy()
min_max_scaler = preprocessing.MinMaxScaler()
scaled = min_max_scaler.fit_transform(normalized_data)
movie_normalized = pd.DataFrame(scaled)
movie_normalized.columns = ['Sonic', 'Bad_Boys', '1917','Jumanji','Joker','User Average']
names = ['Bob','Eugene','Dante','Mike','Jess','Average Movie Rating']
movie_index = pd.DataFrame(names, columns=['a'])
movie_normalized.insert(0,'Names',movie_index['a'])
movie_normalized.set_index(['Names'])
```
